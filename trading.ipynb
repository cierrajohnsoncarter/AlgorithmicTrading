Create the simple moving average with a 30 day window

```
SMA30 = pd.DataFrame()

SMA30['Adj Close Price'] = AAPL['Adj Close'].rolling(window=30).mean()
SMA30
```

Create the simple moving average with an 100 day window

```
SMA100 = pd.DataFrame()
SMA100['Adj Close Price'] = AAPL['Adj Close'].rolling(window=100).mean()
SMA100
```

VIsualize the data

```
plt.figure(figsize=(12.5, 4.5))

plt.plot(AAPL['Adj Close'], label = 'AAPL')
plt.plot(SMA30['Adj Close Price'], label= 'SMA30')
plt.plot(SMA100['Adj Close Price'], label= 'SMA100')

plt.title('Apple Adj. Close Price History')
plt.xlabel('September 09, 2014 - March 03, 2018')
plt.ylabel('Adj. Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```

Create a new dataframe to store the data

```
data = pd.DataFrame()
data['AAPL'] = AAPL['Adj Close']
data['SMA30'] = SMA30['Adj Close Price']
data['SMA100'] = SMA100['Adj Close Price']
data
```

Create a function to determine when to buy and sell the asset/stock

```
def buy_sell(data):
  sigPriceBuy = []
  sigPriceSell = []
  flag = -1 # Tells us when the two moving averages cross eachother
  
  for i in range(len(data)):
    # Buy the stock
    if data['SMA30'][i] > data['SMA100'][i]:
      if flag != 1:
        sigPriceBuy.append(data['AAPL'][i])
        sigPriceSell.append(np.nan)
        flag = 1
      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    # Sell the stock
    elif data['SMA30'][i] < data['SMA100'][i]:
      if flag != 0:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(data['AAPL'][i])
        flag = 0
      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    else:
      sigPriceBuy.append(np.nan)
      sigPriceSell.append(np.nan)

  return (sigPriceBuy, sigPriceSell)
```

Store the buy and sell data in a variable

```
buy_sell = buy_sell(data)
data['Buy_Signal_Price'] = buy_sell[0]
data['Sell_Signal_Price'] = buy_sell[1]
```

Show the data

```
data
```

Visualize the data and the strategy to buy and sell the stock

```
plt.figure(figsize=(12.6, 4.6))

plt.plot(data['AAPL'], label='AAPL', alpha = 0.35)
plt.plot(data['SMA30'], label='SMA30', alpha = 0.35)
plt.plot(data['SMA100'], label='SMA100', alpha = 0.35)

plt.scatter(data.index, data['Buy_Signal_Price'], label = 'Buy', marker = '^', color = 'green')
plt.scatter(data.index, data['Sell_Signal_Price'], label = 'Sell', marker = 'v', color = 'red')

plt.title('Apple Adj Close Price History Buy & Sell Signals')
plt.xlabel('September 09, 2014 - March 03, 2018')
plt.ylabel('Adj Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```
